```python
import pandas as pd
```

```python
puzzles = pd.read_csv("E:\Wszystko\SGH\I rok\II sem\Statistical Learning Methods\puzzles.csv")
```

```python
puzzles.head()
```

```python
puzzles.columns = ["PuzzleId", "FEN", "Moves",
                            "Rating","RatingDeviation",
                            "Popularity", "NbPlays",
                            "Themes","GameUrl"]
```

```python
puzzles.head()
```

```python
import statistics as st
```

```python
plays_lo = st.median(puzzles["NbPlays"])
puzzles = puzzles[puzzles["NbPlays"] >= plays_lo]

puzzles
```

```python
import numpy as np
```

```python
rating_lo = 1500
rating_hi = np.quantile(puzzles["Rating"], 0.99)
puzzles = puzzles[puzzles["Rating"] >= rating_lo]
puzzles = puzzles[puzzles["Rating"] <= rating_hi]
puzzles
```

```python
import matplotlib.pyplot as plot
```

```python
plot.hist(puzzles["Rating"],bins=100) 
#axis([xmin,xmax,ymin,ymax])
plot.xlabel('Weight')
plot.ylabel('Probability')
```

```python
plot.hist(puzzles["Popularity"],bins=100) 
#axis([xmin,xmax,ymin,ymax])
plot.xlabel('Weight')
plot.ylabel('Probability')
```

```python
ratings = np.unique(puzzles["Rating"])
```

```python
mean_popularities = puzzles.groupby('Rating')['Popularity'].mean()
```

```python
mean_popularities
```

```python
plot.scatter(ratings, mean_popularities)
plot.xlabel("rating")
plot.ylabel("mean popularity")
```

```python
from scipy.interpolate import interp1d
import statsmodels.api as sm
```

```python
lowess = sm.nonparametric.lowess(mean_popularities, ratings, frac=.1)
lowess_x = list(zip(*lowess))[0]
lowess_y = list(zip(*lowess))[1]
```

```python
f = interp1d(lowess_x, lowess_y, bounds_error=False)
xnew = [i/10. for i in range(400)]
```
